```python
import math

def is_triangle_number(T):
    k = (-1 + math.sqrt(1 + 8*T)) / 2
    return k % 1 == 0
```

```python
print(is_triangle_number(15))
```

```python
def f(target_count):
    a_prev = 3
    a_curr = 4
    index = 1
    triangle_count = 1  # a_0 = 3 is a triangle
    
    while triangle_count < target_count:
        is_tri = is_triangle_number(a_curr)
        
        if is_tri:
            triangle_count += 1
            if triangle_count == target_count:
                return index
        
        if is_tri:
            a_next = a_curr + 1
        else:
            a_next = 2*a_curr - a_prev + 1
        
        a_prev = a_curr
        a_curr = a_next
        index += 1
```

```python
print(f(10))
```

```python
def analyze_triangle_gaps():
    a_prev = 3
    a_curr = 4
    index = 1
    triangle_count = 1
    last_triangle_index = 0  # a_0 = 3
    
    # Store data about each gap
    gaps = []
    ks = []
    
    while triangle_count < 70:  # Check first 15 triangles
        if is_triangle_number(a_curr):
            triangle_count += 1
            gap = index - last_triangle_index
            gaps.append({
                'triangle_num': triangle_count,
                'value': a_curr,
                'index': index,
                'gap_from_previous': gap
            })
            last_triangle_index = index
        
        # Generate next
        if is_triangle_number(a_curr):
            a_next = a_curr + 1
        else:
            a_next = 2*a_curr - a_prev + 1
        
        a_prev = a_curr
        a_curr = a_next
        index += 1
    
    return gaps

gaps = analyze_triangle_gaps()
for g in gaps:
    print(g)
```

```python
import math

T = 20614324036223469
discriminant = 1 + 8 * T

# Python 3.8+ has integer square root
isqrt_val = math.isqrt(discriminant)

# Check if it's a perfect square
if isqrt_val * isqrt_val == discriminant:
    k = (isqrt_val - 1) // 2
    print(f"k = {k}")
    
    # Verify
    verify = k * (k + 1) // 2
    print(f"T_k = {verify}")
    print(f"Match? {verify == T}")
else:
    print("Not a perfect square - might not be a triangle number!")
```
